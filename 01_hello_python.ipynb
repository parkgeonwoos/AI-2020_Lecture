##For statement

```
A = range(5)
print(A)
print(A[2])
```

```
for i in range(5):
  print(i,A[i])
```

###**Operators**


##**+연산자**

```
print((1,2,3)+(4,5,6))
print([1,2,3]+[4,5,6])
print("Hello"+""+"World")
```

##***연산자**

The *operator produces a new tuple,listt,or string that "repeats" the original content.

```
print((1,2,3)*3)
print([1,2,3]*3)
print("Hello"*3)
```

##Containers

Python includes several built-in container types:lists,dictionaries,sets,and tuples.

##Tuple


A simple immutable(변경할 수 없는,불변의)ordered sequence of items

```
months = ('January','February','March','April','May','June','July','August','September','October','November','December')
 
print(months[0])
print("index of 7 ==> " , months[7])
```

```
for item in months:    
  print (item)
```

```
t = ('john', 32, (2,3,4,5), 'hello') 
print(t)
print(t[2]) 
print(t[2][1])
print(t[:2]) # index 포함 X print(t[2:]) # index 포함 O
 
print(t[-1]) 
print(t[-2])
```

##List


Mutable(바꿀수 있는,변경가능한)ordered sequence of itomes of mixed types

```
li = ['hallym', 1, 3.141572, 'hello'] 
print(li) 
li[1] = 45 
print(li) 
li.append('September') 
print(li)
```

```
v = [] 
for i in range(0,3):    
  v.append(i)    
  print(v)
```
