##Sigmoid


A sigmoid function is a mathematical function having a characteristic "S"-shaped curve or sigmoid curve.


$f(x) = \cfrac{1}{1+e^{-x}}$


```
def sigmoid(x):    
  return 1 / (1 +np.exp(-x))
```

```
x = np.arange(-5.0, 5.0, 0.1) 
y2 = sigmoid(x)
```

```
plt.plot(x, y2)
plt.grid() 
plt.show()
```

$f(x) = \cfrac{1}{1+e^{-c_{1}x}}$

```
def sigmoid2(x, c1=1):    
  return 1 / (1 +np.exp(-c1 * x))
```

```
y2 = sigmoid2(x, 2)
```

```
plt.plot(x, y2) 
plt.grid()
```

```
for i in range(10):  
  c1 = 0.2 * (i+1)  
  y2 = sigmoid2(x, c1)  
  plt.plot(x, y2) 
plt.grid()
```

$f(x) = \cfrac{1}{1+e^{-c_{1}(x-c_{2})}}$

```
def sigmoid3(x, c1=1, c2=0):    
  return 1 / (1 +np.exp(-c1 * (x- c2)))
```

```
y3 = sigmoid3(x)
```

```
plt.plot(x, y3) 
plt.grid()
```

```
for i in range(10):  
  c2 = 0.5 * i  
  c1 = 2
  y3 = sigmoid3(x, c1, c2)
  plt.plot(x, y3)
plt.grid()
```
