```
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import numpy as np
```

```
tpts = np.array([[1,2],[1,5],[2,2],[3,3.5],[7,5],[9,4],[8,2]])
labels = [1,1,1,1,0,0,0]
```

```
tpts
```

```
tpts.shape
```

```
plt.scatter(tpts[:,0],tpts[:,1],c=labels)
```

```
clf = SVC(kernel='linear')
```

```
%%time
clf.fit(tpts,labels)
```

##테스트해보기


```
pred1 = clf.predict([[7,4]])
print(pred1[0])
```

```
pred1 = clf.predict([[3,3]])
print(pred1[0])
```

```
pred1 =clf.predict([[7,4.1]])
print(pred1[0])
```

##그래프 그려보기


```
plt.scatter(tpts[:,0],tpts[:,1],c=labels)
```

```
clf.coef_
```

```
clf.intercept_
```

```
def getXY(clf):
  w=clf.coef_[0]
  i=clf.intercept_[0]
  print('weights=',w)
  print('intercept_=',i)
  a=-w[0]/w[1]
  b=-i/w[1]
  x=np.linspace(1,9)
  y=a*x+b
  return x,y
```

```
x,y=getXY(clf)
```

```
x
```

```
plt.scatter(tpts[:,0],tpts[:,1],c=labels)
plt.plot(x,y,'-')
```
